```python
import os
import random
import numpy as np
```

```python
key = np.arange(0, 250)
key
```

```python
random.shuffle(key)
```

```python
key
```

```python
key_label = []
for j in range(0, 250):
    if j < 125:
        key_label.append('melanoma')
    else:
        key_label.append('not melanoma')
        
```

```python
key_full = np.vstack((key, key_label)).T
key_full
```

```python
np.savetxt('keys_for_melanoma_250.txt', key_full, fmt='%s')
```

```python
i = 0
for file in os.listdir('melanoma_250'):
    os.rename(f'melanoma_250/{file}', f'melanoma_250/{key[i]}_{file}')
    i = i + 1
```
